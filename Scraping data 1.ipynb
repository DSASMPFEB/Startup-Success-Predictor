```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

# Launch browser with auto-managed driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.startupindia.gov.in/content/sih/en/search.html")
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
cards = soup.select('.category-card.search-card.new-eco-card')
print("Total startup cards found:", len(cards))
```

```python
startups = []

for card in soup.select('.category-card.search-card.new-eco-card'):
    try:
        name = card.select_one('.events-details h3').text.strip()
        stage = card.select_one('.highlighted-text').text.strip()
        location_tags = card.select('li.location span')
        location = ', '.join([tag.text.strip() for tag in location_tags])
        sector = card.select_one('.dept').text.strip()
        

        founded = None
        inc_date = card.select_one('.incorporation-date')
        if inc_date:
            founded = inc_date.text.strip()

        funding = None
        funding_text = card.select_one('.funding-info')
        if funding_text:
            funding = funding_text.text.strip()

        startups.append({
            'name': name,
            'stage': stage,
            'location': location,
            'sector': sector,
            'founded': founded,
            'funding': funding
        })

    except Exception as e:
        print("Error parsing card:", e)
```

```python
import pandas as pd
```

```python
df_startup_india = pd.DataFrame(startups)
```
