```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://yourstory.com/companies/search?q=india&page=1")
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.CLASS_NAME, "hit"))
)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
startup_data = []

for row in soup.select("tr.hit"):
    cells = row.find_all("td")
    if len(cells) >= 6:
        name = cells[1].get_text(strip=True)
        sectors = cells[2].get_text(strip=True)
        headquarters = cells[3].get_text(strip=True)
        funding = cells[4].get_text(strip=True)
        employee_count = cells[5].get_text(strip=True)

        startup_data.append({
            "name": name,
            "sectors": sectors,
            "headquarters": headquarters,
            "funding": funding,
            "employee_count": employee_count
        })
```

```python
import pandas as pd

df = pd.DataFrame(startup_data)
print(df.head())
```

```python
df.to_csv("yourstory_startups.csv", index=False)
```

```python
df = pd.read_csv("yourstory_startups.csv")
```

```python
df
```

```python
import pandas as pd
import glob

# Step 1: Collect all CSV file paths
csv_files = sorted(glob.glob("yourstory_startups*.csv"))
```

```python
# Step 2: Read and store DataFrames
dfs = [pd.read_csv(file) for file in csv_files]
```

```python
# Step 3: Concatenate all DataFrames
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
merged_df
```

```python
merged_df.duplicated().value_counts()
```

```python
# Step 4: Drop duplicates (based on all columns)
merged_df.drop_duplicates(inplace=True)
```

```python
# Step 5: Save the final merged dataset
merged_df.to_csv("yourstory_startups_merged.csv", index=False)

print(f"Merged {len(csv_files)} files into one dataset with {len(merged_df)} unique rows.")
```

```python
# Define funding round categories
funding_map = {
    
    'OK India': 'Seed',
    'SweeDesi India':'Seed',
    'One Connected': 'Seed',
    'Esports Daily': 'Series A',  
    'Eduncle': 'Series A',  
    'Buyceps': 'Seed',
    'eBik': 'Seed',
    'Bolo Indya': 'Seed',
    'Vikalp India': 'Series A',
    'US India Blockchain Council': 'Series A',
    'Mapmygenome': 'Series A',
    'Eko': 'Other',
    'GIVA Jewellery': 'Series B',
    'NbliK': 'Angel',
    'IndicWisdom': 'Equity Crowdfunding',
    'Trylaa': 'Pre Seed'
}

# Assign funding round
data['funding_round'] = data['name'].map(funding_map)
```
