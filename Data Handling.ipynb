```python
import re
import numpy as np
import pandas as pd

def extract_primary_city(val):
    if pd.isna(val):
        return np.nan
    val = str(val).lower().strip()

    # Remove bracketed content: e.g. "Mumbai (HQ)" → "Mumbai"
    val = re.sub(r"\(.*?\)", "", val)

    # Remove content after '&': e.g. "Delhi & NCR" → "Delhi"
    val = val.split("&")[0]

    # Split on '/' and take the first part: e.g. "SFO / Bangalore" → "SFO"
    parts = re.split(r"\s*/\s*", val)
    return parts[0].strip() if parts else val.strip()
```

```python
data['city'] = data['city'].apply(extract_primary_city)
```

```python
def convert_numeric_round(val):
    numeric_map = {
        1: "pre-seed",
        2: "seed",
        3: "series a",
        4: "series b",
        5: "series d",
        6: "series e",
        7: "growth",
        8: "ipo",
        9: "acquisition",
        10: "series f",
        11: "series g",
        12: "series h",
        13: "series j",
        14: "series x",
        15: "venture",
        16: "private equity",
        17: "bridge",
        18: "equity",
        20: "debt",
        24: "grant"
    }
    try:
        val = int(float(val))  # handles both int and float strings
        return numeric_map.get(val, str(val).lower().strip())
    except:
        return str(val).lower().strip()
```

```python
def normalize_round(raw):
    raw = str(raw).lower().strip()
    if 'pre-seed' in raw: return 'pre-seed'
    elif 'seed' in raw: return 'seed'
    elif 'angel' in raw: return 'angel'
    elif 'series a' in raw: return 'series a'
    elif 'series b' in raw: return 'series b'
    elif 'series c' in raw: return 'series c'
    elif 'series d' in raw: return 'series d'
    elif 'series e' in raw: return 'series e'
    elif 'series f' in raw: return 'series f'
    elif 'series g' in raw: return 'series g'
    elif 'series h' in raw: return 'series h'
    elif 'series j' in raw: return 'series j'
    elif 'bridge' in raw: return 'bridge'
    elif 'growth' in raw: return 'growth'
    elif 'venture' in raw: return 'venture'
    elif 'private equity' in raw: return 'private equity'
    elif 'debt' in raw: return 'debt'
    elif 'equity' in raw: return 'equity'
    elif 'grant' in raw: return 'grant'
    elif raw in ['-', '–', '—', 'not disclosed', 'undisclosed round']: return 'unknown'
    elif raw.isdigit(): return 'numeric'
    else: return 'other'


data['funding_round'] = data['funding_round'].apply(normalize_round)
```

```python
import re
import numpy as np
import pandas as pd

def parse_funding(val):
    try:
        if pd.isna(val):
            return np.nan
        if isinstance(val, (int, float)):
            return float(val)
        val = str(val).lower().replace(",", "").strip()
        match = re.match(r"\$?([\d\.]+)\s*(million|billion)?", val)
        if match:
            num = float(match.group(1))
            unit = match.group(2)
            if unit == "million":
                return num * 1e6
            elif unit == "billion":
                return num * 1e9
            else:
                return num
        return np.nan
    except:
        return np.nan
```
